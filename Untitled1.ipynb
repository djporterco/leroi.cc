```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
def candles(exchange, quoteTradingSymbol, baseTradingSymbol, interval):
    response = requests.get(f"https://dev-api.shrimpy.io/v1/exchanges/{exchange}/candles?quoteTradingSymbol={quoteTradingSymbol}&baseTradingSymbol={baseTradingSymbol}&interval={interval}")
    return json_normalize(response.json())
```

```python
df1 = candles('binance', 'btc', 'eth', '6h').head()
```

```python
df1.to_csv('candles/eth_candles.csv', encoding='utf-8', index=False)

#df1.to_csv('candles/eth_candles.csv')
```

```python
def ticker(exchange):
    response = requests.get(f"https://dev-api.shrimpy.io/v1/exchanges/{exchange}/ticker")
    df = json_normalize(response.json())
    return df.set_index('lastUpdated')
```

```python
ticker('binance').head()
```

```python
def list_exchanges():
    res = requests.get("https://dev-api.shrimpy.io/v1/list_exchanges")
    return json_normalize(res.json())
```

```python
list_exchanges()
```

```python
df = ticker('binance')
```

```python
coins = df['symbol'].tolist()
```

```python
intervals = ['1m', '5m', '15m', '1h', '6h', '1d']
```

```python
for interval in intervals:
    for coin in coins:
        df = candles('binance', 'btc', coin, interval)
        df.to_csv('candles/binance_' +  interval  + '_BTC_' + coin + '_candles.csv', encoding='utf-8', index=False)
```
